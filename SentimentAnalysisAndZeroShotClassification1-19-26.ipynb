### Using the tokenizer and model separately

##### When to use specific classes:
- You know the model architecture (e.g., you're using only BERT).
- You need access to model-specific methods or behaviors.
- You're extending or modifying the model for research.

Otherwise, the Auto* classes are preferred for general-purpose applications and code that needs to support multiple model types.

```python
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification
from transformers import AutoModelForSequenceClassification, AutoTokenizer
```

```python
tokenizer = DistilBertTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")

tokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")

tokenizer
```

```python
model = DistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")

model
```

```python
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")

model
```

```python
inputs = tokenizer("The constant delays are making it difficult for us to stay motivated on this project.", return_tensors="pt")

inputs
```

```python
logits = model(**inputs).logits

logits
```

Using a negative value for the axis counts dimensions from the end of the shape, rather than the beginning. Specifically, axis=-1 refers to the last dimension of the tensor.

```python
predicted_class_id = int(logits.argmax(axis=-1)[0])

print(f"Predicted class id {predicted_class_id}")

print(f"Label {model.config.id2label[predicted_class_id]}")
```

## Zero-shot Classification

- Go to https://huggingface.co/ -> Models
- Click on Zero-shot Classification on the left pane
- Search for facebook/bart-large-mnli
- Click on the model and show the details

```python
zeroshot_pipeline = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

zeroshot_pipeline
```

```python
text = "The fog crept silently through the darkened streets, swallowing the last remnants of daylight as the town braced for the unknown."

results = zeroshot_pipeline(text, candidate_labels=['romance', 'science fiction', 'horror', 'young adult', 'comedy'])

results
```

Note that the probabilities sum up to a 100%

```python
import pandas as pd

df = pd.DataFrame({'labels': results['labels'], 'scores': results['scores']})

df
```

```python
text = "The company's quarterly earnings report exceeded analysts' expectations, driving a significant uptick in stock prices."

results = zeroshot_pipeline(text, candidate_labels=['technology', 'politics', 'entertainment', 'climate change', 'business'])

df = pd.DataFrame({'labels': results['labels'], 'scores': results['scores']})

df
```

Note that the probabilities do not sum up to 1

```python
text = "The proposed legislation aims to address climate change by incentivizing green energy production, but it has sparked fierce debate among lawmakers over its economic impact."

results = zeroshot_pipeline(text, candidate_labels=['technology', 'climate change', 'politics', 'economics', 'finance', 'business', 'environment'], multi_label=True)

df = pd.DataFrame({'labels': results['labels'], 'scores': results['scores']})

df
```

```python
text = "As the ancient clock struck midnight, she uncovered the dusty tome that had been hidden away for centuries, its pages whispering secrets of a forgotten world."

results = zeroshot_pipeline(text, candidate_labels=['gothic', 'romance', 'politics', 'comedy', 'finance', 'mystery'], multi_label=True)

df = pd.DataFrame({'labels': results['labels'], 'scores': results['scores']})

df
```
