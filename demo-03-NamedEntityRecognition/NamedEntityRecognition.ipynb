```python
from transformers import pipeline
```

https://huggingface.co/dslim/bert-large-NER

- O	Outside of a named entity
- B-MISC	Beginning of a miscellaneous entity right after another miscellaneous entity
- I-MISC	Miscellaneous entity
- B-PER	Beginning of a person’s name right after another person’s name
- I-PER	Person’s name
- B-ORG	Beginning of an organization right after another organization
- I-ORG	organization
- B-LOC	Beginning of a location right after another location
- I-LOC	Location

```python
ner_pipeline = pipeline("ner", model="dslim/bert-large-NER")

ner_pipeline
```

```python
text = "Jane will be at the the United Nations General Assembly will be held in New York City on September 25, 2025."

print(text)

ner_pipeline(text)
```

aggregation_strategy is used to group the entities in the predictions. It will group "New York City" together

```python
texts = [
    "The Eiffel Tower, an iconic symbol of Paris and French culture, drawing millions of visitors who come to admire its iron lattice structure and breathtaking views of the city.",
    "Mount Everest is the highest mountain in the world, towering over the Himalayas and straddling the border between Nepal and Tibet."
]

print(texts)

ner_pipeline(texts, aggregation_strategy="simple")
```

```python
ner_results = ner_pipeline(texts, aggregation_strategy="simple")

for idx, (text, result) in enumerate(zip(texts, ner_results), 1):
    print(f"\nText {idx}: {text}")

    if result and isinstance(result[0], dict):
        print("Entities:")
        for entity in result:
            for key, value in entity.items():
                print(f"  - {key}: {value}")
```

```python
from transformers import BertForTokenClassification, BertTokenizer, AutoTokenizer,AutoModelForTokenClassification
```

```python
FB_MODEL = "xlm-roberta-large-finetuned-conll03-english"

FB_MODEL
```

```python
tokenizer = AutoTokenizer.from_pretrained(FB_MODEL)

tokenizer
```

The original model was trained using PyTorch and we convert it to TensorFlow using from_pt = True

```python
model = AutoModelForTokenClassification.from_pretrained(FB_MODEL)

model
```

```python
text = """
The Grand Canyon is a vast natural wonder located in Arizona, USA. It was shaped by the Colorado River over millions of years,
creating stunning geological formations and rich biodiversity. The Grand Canyon National Park is managed by the National Park Service, and
renowned geologist John Wesley Powell was among the first to document the canyon's remarkable features in the late 19th century.
"""

text
```

```python
tokens = tokenizer.encode_plus(text, return_tensors="pt")

tokens
```

```python
outputs = model(input_ids=tokens["input_ids"], attention_mask=tokens["attention_mask"])

outputs
```

##### obtain predicted labels by finding the index with the highest probability

```python
import torch

predicted_labels = torch.argmax(outputs.logits, dim=2).numpy()

predicted_labels
```

##### converting back

```python
model.config.id2label
```

```python
predicted_tokens = tokenizer.convert_ids_to_tokens(tokens["input_ids"][0])

predicted_tokens
```

```python
for token, label in zip(predicted_tokens, predicted_labels[0]):
    print(f"{token} - {model.config.id2label[label]}")
```

##### removing all the unnessary _, O and parsing it properly for better understanding

```python
result_dict = {}

current_entity = ""
current_label = None

for token, label_id in zip(predicted_tokens, predicted_labels[0]):
    label = model.config.id2label[label_id]

    # Split I-PER, B-PER etc into two parts and access the second part
    # For O or outside just have the label O
    entity_type = label.split('-', 1)[1] if '-' in label else 'O'

    if entity_type != 'O':
        # For XLM-Roberta, subword tokens are prefixed with '▁' (U+2581)
        # Anything with '_' is the start of a word
        if token.startswith("▁"):
            if current_entity:
                result_dict[current_entity] = current_label
            current_entity = token[1:]  # Remove the '▁' at the start
        else:
            current_entity += token  # Append subword without space
        current_label = entity_type
    elif current_entity:
        result_dict[current_entity] = current_label
        current_entity = ""
        current_label = None

# Handle the last entity if it exists
if current_entity:
    result_dict[current_entity] = current_label

print(result_dict)
```

```python
for entity, label in result_dict.items():
    print(f"{entity} - {label}")
```
